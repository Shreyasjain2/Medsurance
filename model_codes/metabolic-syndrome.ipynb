```python
import pandas as pd
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras import layers
import kagglehub
# from sklearn.preprocessing import OneHotEncoder

# import pandas as pd
# Download dataset
path = kagglehub.dataset_download("antimoni/metabolic-syndrome")

# Load CSV file (assuming a single CSV file in the downloaded path)
import os
df = pd.read_csv(os.path.join(path, "Metabolic Syndrome.csv"))

# Select relevant features
features = ["Age","Sex", "WaistCirc", "BMI", "Albuminuria", "UrAlbCr", "UricAcid", "BloodGlucose", "HDL", "Triglycerides"]
target = "MetabolicSyndrome"

df = df[features + [target]].dropna()  # Drop missing values

# Split dataset into features and target
X = df[features]
y = df[target].values
from sklearn.preprocessing import LabelEncoder

label_encoder = LabelEncoder()
X['Sex'] = label_encoder.fit_transform(X['Sex'])  # Assuming 'Gender' is the problematic column


# Normalize data
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Split into training and test sets
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.2, random_state=42)

# Define ANN model
model = keras.Sequential([
    layers.Dense(128, activation='relu', input_shape=(X_train.shape[1],)),
    layers.Dense(64, activation='relu'),
    layers.Dropout(0.2),
    layers.Dense(32, activation='relu'),
    layers.Dropout(0.1),
    layers.Dense(1, activation='sigmoid')  # Binary classification
])

# Compile model
model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])

# Train model
model.fit(X_train, y_train, epochs=30, batch_size=32, validation_data=(X_test, y_test))

# Evaluate model
test_loss, test_acc = model.evaluate(X_test, y_test)
print(f"Test Accuracy: {test_acc:.4f}")
```
